```python
import os
os.chdir("..")

import pandas as pd
import matplotlib.pyplot as plt

raw = pd.read_csv("data/processed/macro_panel_raw_units.csv", index_col=0, parse_dates=True)
scaled = pd.read_csv("data/processed/macro_panel.csv", index_col=0, parse_dates=True)

print("Shape:", raw.shape)
print("\nRaw units summary:")
print(raw.describe())
fig, axes = plt.subplots(5, 1, figsize=(12, 14))
for ax, col in zip(axes, raw.columns):
    ax.plot(raw.index, raw[col], linewidth=1)
    ax.set_title(col, fontsize=12, fontweight="bold")
    ax.axhline(0, color="gray", linestyle="--", linewidth=0.7)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("outputs/figures/02_processed_panel.png", dpi=150)
plt.show()
```
